```python
from nltk.corpus import stopwords
stopwords.words('english')[0:10]
```

```python
def text_clean(text):
    
    nopunc = [char for char in text if char not in string.punctuation]
    
    nopunc = ''.join(nopunc)
    
    return [word for word in nopunc.split() if word.lower() not in stopwords.words('english')]
```

```python
df['Header text:'].head(5)
```

```python
df['Text Length'] = df['Header text:'].apply(text_clean).apply(len)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
bowxform = CountVectorizer(analyzer=text_clean).fit(df['Header text:'])
```

```python
print(len(bowxform.vocabulary_))
```

```python
mess3 = df['Header text:'][2]
print(mess3)
```

```python
bow3 = bowxform.transform([mess3])
print(bow3)
```

```python
print(bowxform.get_feature_names()[9909])
```

```python
bow = bowxform.transform(df['Header text:'])
```

```python
print(bow.shape)
print(bow.nnz) # amount of non-zeros
```

```python
sparsity = (100.0 * bow.nnz / (bow.shape[0] * bow.shape[1]))
print('sparsity: {}'.format(round(sparsity)))
```
